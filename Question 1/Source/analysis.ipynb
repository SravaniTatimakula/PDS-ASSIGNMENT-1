```python
from google.colab import files
data_to_load = files.upload()
```

```python
import io
import pandas as pd
from scipy import stats
df = pd.read_csv(io.BytesIO(data_to_load['raw_handgrip_data.csv']))
```

```python
df.head()
```

```python
Clean_handgrip_data = df.dropna()

# Print the modified dataframe
print(Clean_handgrip_data)
```

```python
Clean_handgrip_data.to_csv('Clean_handgrip_data.csv', index=False)
```

```python
# Separate the data into two groups based on Frailty and grip strength
group1 = Clean_handgrip_data[Clean_handgrip_data['Frailty'] == 'N']['Grip_strength']
group2 = Clean_handgrip_data[Clean_handgrip_data['Frailty'] == 'Y']['Grip_strength']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=False)

# Print the results
print('T-statistic:', t_statistic)
print('P-value:', p_value)
```
